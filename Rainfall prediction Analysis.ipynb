```python
# importing the libraries.
import numpy as np
import pandas as pd
import sklearn as sk
import sklearn.linear_model as LinearRegression
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("austin_weather_final.csv")
data
```

```python
x = data.drop(["PrecipitationSumInches"], axis=1)
```

```python
y = data["PrecipitationSumInches"]
```

```python
# reshaping in 2D vector.
y = y.values.reshape(-1,1)
```

```python
y
```

```python
day_index = 798
days =[i for i in range(y.size)]
days
```

```python
from sklearn.linear_model import LinearRegression



# Initialize the linear regression classifier
clf = LinearRegression()

# Train the classifier
clf.fit(x, y)
```

```python
# plot a graph
print("The precipitation trend Graph")
plt.scatter(days,y, color='g')
plt.scatter(days[day_index], y[day_index], color='r')
plt.title("Precipitation Level")
plt.xlabel("Days")
plt.ylabel("Precipitation in inches")
plt.show()
x_vis = x[["TempAvgF", "HumidityAvgPercent", "DewPointAvgF", "SeaLevelPressureHighInches", "VisibilityAvgMiles"]]
```

```python
import matplotlib.pyplot as plt

print("The Precipitation vs attributes trend graph:")

colors = ['b', 'g', 'r', 'c', 'm', 'y']

for i in range(x_vis.columns.size):
    plt.subplot(3, 2, i+1)
    plt.scatter(days, x_vis[x_vis.columns.values[i][:100]], color=colors[i % len(colors)])
    plt.scatter(days[day_index], x_vis[x_vis.columns.values[i]][day_index], color='r', label='Chosen Day')
    plt.title(x_vis.columns.values[i])
    plt.xlabel('Days')
    plt.ylabel('Attribute Value')
    plt.tight_layout()

plt.legend()
plt.show()


```

```python
y_pred = clf.predict(x)
```

```python
# Plot the actual vs. predicted PrecipitationSumInches
plt.scatter(days, y, color='b', label='Actual')
plt.scatter(days, y_pred, color='r', label='Predicted')
plt.scatter(days[day_index], y[day_index], color='g', label='Chosen Day')
plt.title('Actual vs. Predicted PrecipitationSumInches')
plt.xlabel('Days')
plt.ylabel('PrecipitationSumInches')
plt.legend()
plt.show()
```

```python
# Plot trend lines
plt.plot(days, y, label='Actual', color='b')
plt.plot(days, y_pred, label='Predicted', color='r')
plt.scatter(days[day_index], y[day_index], color='g', label='Chosen Day')
plt.title('Trend Lines of Actual and Predicted PrecipitationSumInches')
plt.xlabel('Days')
plt.ylabel('PrecipitationSumInches')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Calculate Mean Squared Error (MSE) and R-squared
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f'Mean Squared Error (MSE): {mse}')
print(f'R-squared: {r2}')
```

```python

plt.scatter(y, y_pred, color='b')
plt.plot([min(y), max(y)], [min(y), max(y)], linestyle='--', color='r', label='Ideal Line')
plt.title('Actual vs. Predicted PrecipitationSumInches')
plt.xlabel('Actual PrecipitationSumInches')
plt.ylabel('Predicted PrecipitationSumInches')
plt.legend()
plt.show()
```
